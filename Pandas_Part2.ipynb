```python
df=pd.read_excel('CustomerChurn.xlsx')
```

# MissingValueImputation

```python
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    df.loc[df.sample(frac=0.01).index,col]=np.nan
```

```python
df.isnull().sum()
```

## method-1

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.shape
```

## method-2

```python
df.isnull().sum()
```

```python
df['Customer ID']=df['Customer ID'].replace(to_replace='np.nan',
                         value='Missing')
```

```python
df[df['Customer ID']=='Missing'].shape
```

## method-3

```python
df['LoyaltyID']=df.LoyaltyID.fillna(value='SAGESGAESGAE')
```

```python
df[df['LoyaltyID']=='SAGESGAESGAE']
```

```python
np.round(df.Tenure.mean())
```

```python
df['Tenure']=df.Tenure.fillna(value=np.round(df.Tenure.mean()))
```

## method-6

```python
df.info()
```

```python
for fea in df.columns:
    if df[fea].dtype=='O':
        df[fea]=df[fea].replace(to_replace=np.nan,
                           value='Missing') 
    else:
        df[fea]=df[fea].fillna(value=np.round(df[fea].mean()))
```

```python
df[df['Senior Citizen']=='Missing']
```

## num2Cat

```python
df1=df[['Customer ID','Tenure']]
df1.head()
```
